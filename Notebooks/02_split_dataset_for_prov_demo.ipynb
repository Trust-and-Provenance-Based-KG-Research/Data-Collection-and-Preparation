```python
import pandas as pd
import math
import os
```

```python
# Data Paths
input_path = "../Prove_Data/EDA_Datasets/Video_Metadata.csv"

output_dir = "../Prove_Data/Provenance_Datasets"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)
```

```python

# Load the dataset
df = pd.read_csv(input_path)
total_rows = len(df)
batch_size = math.ceil(total_rows / 4)

```

```python
# Split and save
for i in range(4):
    start = i * batch_size
    end = start + batch_size
    batch_df = df.iloc[start:end]

    batch_filename = f"{['first', 'second', 'third', 'fourth'][i]}_batch_metadata.csv"
    output_path = os.path.join(output_dir, batch_filename)

    batch_df.to_csv(output_path, index=False)
    print(f"Saved: {output_path} ({len(batch_df)} rows)")
```
